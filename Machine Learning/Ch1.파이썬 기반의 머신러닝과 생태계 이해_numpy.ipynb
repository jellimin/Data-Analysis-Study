## ■ 머신러닝의 개념

* 머신러닝이란 애플리케이션을 수정하지 않고도 데이터를 기반으로 패턴을 학습하고 결과를 추론하는 알고리즘 기법을 통칭

## ■ 넘파이 배열 ndarray 소개
* Numpy : Numerical Python, 파이썬에서 선형대수 기반의 프로그램을 쉽게 만들 수 있도록 지원하는 대표적인 패키지
* ndarray : 넘파이 기반의 데이터 타입, N차원 배열 객체
* 1차원 배열은 선의 개념, 2차원 배열은 면의 개념, 3차원 배열은 공간의 개념
<br>
* Numpy 모듈의 `array()` 함수로 생성
* 인자로 주로 파이썬 list 또는 ndarray 입력

### ndaaray 형태(shape)와 차원
 * ndarray의 shape는 `ndarray.shape` 속성으로, 차원은 `ndarray.ndim` 속성으로 알 수 있다.
 * shape 속성은 ndarray의 크기, 즉 행과 열의 수를 튜플 형태로 가지고 있음
|array|차원|shape|
|:---:|:---:|:---:|
|[1 2 3]|1차원|(3,)|
|[[1 2 3]<br>[4 5 6]]|2차원|(2, 3)|

```python
import numpy as np
```

```python
# array 생성 및 타입, shape

list1 = [1, 2, 3]
print('list type :', type(list1), '\n')

array1 = np.array(list1) # 1차원 array
print('array1 type :', type(array1))
print('array1 array 형태 :', array1.shape, '\n') # (3,)를 반환, 콤마(,)가 들어갔으므로 1차원

array2 = np.array([[1, 2, 3],
                   [2, 3, 4]]) # 2차원 array
print('array2 type :', type(array2))
print('array2 array 형태 :', array2.shape, '\n') # (2, 3)을 반환

array3 = np.array([[1, 2, 3]]) # 2차원 array
print('array3 type', type(array3))
print('array3 array 형태', array3.shape, '\n') # (1, 3)을 반환

# array의 차원 : ndim

print('array1 : {0}차원, array2 : {1}차원, array3 : {2}차원'.format(array1.ndim, array2.ndim, array3.ndim))
```

### ndarray 타입
* ndarray 내의 데이터 값은 숫자 값, 문자열 값, 불 값 등이 모두 가능하다.<br>
*하지만, 중요한 것은 ndarray 내의 데이터 타입은 그 연산의 특성상 `같은 데이터 타입`만 가능하다. 즉, 한 개의 ndarray 객체에 int와 float가 함께 있을 수 없다.<br>
* ndarray 내의 데이터 타입은 `ndarray.dtype`으로 확인할 수 있다.

```python
# ndarray의 데이터 타입 : dtype

list1 = [1, 2, 3]
print(type(list1))

array1 = np.array(list1)
print(type(array1))
print(array1, array1.dtype) # dtype은 ndarray 내의 값의 데이터 타입을 알 수 있음
```

### ndaary 타입 변환
* 변경을 원하는 타입을 `ndarray.astype()`에 인자로 입력
* 대용량 데이터 다룰 시 메모리 절약을 위해서 형변환을 특히 고려해야 한다.

### 넘파이 ndarray의 axis 축
* ndarray의 shape는 행, 열, 높이 단위로 부여되는 것이 아니라 axis0, axis1, axis2와 같이 axis 단위로 부여된다.
* axis0은 행 방향성으로 있는 것, axis1은 열 방향성으로 있는 것
<br>
* 1차원 배열은 axis0 밖에 없음
* 2차원 배열의 경우 행은 axis0, 열은 axis1
* 즉, 가장 큰 요소에서 세부 요소로 축 번호가 지정됨


___

```python
# ndarray 내의 데이터 타입이 다른 데이터를 넣었을 때?

list2 = [1, 2, 'test'] # 정수형과 문자열이 혼합
array2 = np.array(list2)
print(array2, array2.dtype) # 문자열(<U11)로 자동 형변환

list3 = [1, 2, 3.0] # 정수형과 실수형이 혼합
array3 = np.array(list3)
print(array3, array3.dtype) # 실수형으로 자동 형변환
```

```python
# ndarray 데이터 타입 변환
array_int = np.array([1, 2, 3])
array_float = array_int.astype('float64') # array_int.astype(np.float64)와 동일
print(array_float, array_float.dtype)

array_int1 = array_float.astype('int32')
print(array_int1, array_int1.dtype)

array_float1 = np.array([1.1, 2.1, 3.1])
array_int2 = array_float1.astype('int32')
print(array_int2, array_int2.dtype)
```

## ■ 넘파이 배열 ndarray 초기화 방법과 ndarray 차원과 크기를 변경하는 reshape()의 이해
* 특정 크기와 차원을 가진 ndarray를 연속값, 0 또는 1로 초기화 생성해야 할 경우 차례대로 `np.arange()`, `np.zeros()`, `np.ones()`를 이용해 쉽게 ndarray를 생성할 수 있다.
* np.zeros()는 만들어야 하는 array의 shape은 아는데 값을 모를 때 자주 사용한다.
* `reshape()`는 ndarray를 특정 차원 및 형태로 변환한다. 변환 형태를 함수 인자로 부여하면 된다.
* `reshape(-1, 5)`와 같이 인자에 -1을 부여하면 -1에 해당하는 axis의 크기는 가변적이되 -1이 아닌 인자값에 해당하는 axis 크기를 인자값으로 고정하여 ndarray의 shape를 반환한다.
* reshape()는 `reshape(-1, 1)`, `reshape(-1,)`과 같은 형식으로 변환이 요구되는 경우가 많다. 주로 머신러닝 API의 인자로 1차원 ndarray를 명확하게 2차원 ndarray로 변환하여 입력하기를 원하거나, 또는 반대의 경우가 있을 경우 reshape()를 이용하여 ndarray의 형태를 변환시켜 주는데 사용된다
 * reshape(-1, 1) : 2차원으로 변환하되 컬럼 axis 크기가 1로 고정
 * reshape(-1,) : 1차원으로 변환

___

```python
# ndarray를 편리하게 생성하기 - arange, zeros, ones

# arange()
sequence_array = np.arange(10)
print(sequence_array)
print(sequence_array.dtype, sequence_array.shape, '\n')

# zeros()
zero_array = np.zeros((3, 2), dtype = 'int32') # shape를 튜플 형태로 지정
print(zero_array)
print(zero_array.dtype, zero_array.shape, '\n')

# ones()
one_array = np.ones((3, 2)) # dtype의 디폴트는 'float64'
print(one_array)
print(one_array.dtype, one_array.shape)
```

```python
# ndarray의 차원과 크기를 변경하는 reshape()
# zeros, ones와 달리 reshape는 튜플로 묶어주어도 되고, 안 묶어도 됨

array1 = np.arange(10)
print('array1 :\n', array1)

# (2, 5) shape으로 변환
array2 = array1.reshape(2, 5)
print('array2 :\n', array2)

# (5, 2) shape으로 변환
array3 = array1.reshape(5, 2)
print('array3 :\n', array3)
```

```python
array1.reshape(4, 3) # shape가 맞지 않아 error가 발생
```

```python
array1 = np.arange(10)
print(array1)

array2 = array1.reshape(-1, 5) # 행의 개수를 가변적
print('array2 shape :', array2.shape)

array3 = array1.reshape(5, -1) # 열의 개수를 가변적
print('array3 shape :', array3.shape)
```

```python
array1 = np.arange(10)
array4 = array1.reshape(-1, 4) # shape가 맞지 않아 error가 발생
```

```python
# 3차원 array의 이용

array1 = np.arange(8)
array3d = array1.reshape((2, 2, 2))
print('array3d :\n', array3d.tolist(), '\n')

# 3차원 ndarray를 2차원 ndarray로 변환하되 칼럼갯수는 ?
array5 = array3d.reshape(-1, 1)
print('array5 :\n', array5.tolist())
print('array5 shape :', array5.shape, '\n')

# 1차원 ndarray를 2차원 ndarray로 변환하되 칼럼갯수는 ?
array6 = array1.reshape(-1, 1)
print('array6 :\n', array6.tolist())
print('array6 shape :', array6.shape, '\n')
```

```python
# 3차원 array를 1차원으로 변환
array1d = array3d.reshape(-1,)
print(array1d)
```

## ■ 넘파이 ndarray 인덱싱 이해 - 단일값, 슬라이싱, 팬시, 불린 인덱싱
|인덱싱 유형|설명|
|:---:|:---:|
|특정 위치의 단일값 추출|원하는 위치의 인덱스 값을 지정하면 해당 위치의 데이터가 반환|
|슬라이싱(Slicing)|연속된 인덱스상의 ndarray를 추출하는 방식|
|팬시 인덱싱(Fancy Indexing)|일정한 인덱싱 집합을 리스트 또는 ndarray 형태로 지정해 해당 위치에 있는 ndarray를 반환|
|불린 인덱싱(Boolean Indexing)|특정 조건에 해당하는지 여부인 True/False 값 인덱싱 집합을 기반으로 True에 해당하는 인덱스 위치에 있는 ndarray를 반환|<br>

* 단일값 추출, 1차원 ndarray : array[index]
* 단일값 추출, 2차원 ndarray : array[rindex, cindex] 
* 슬라이싱, 1차원 ndarray : array[start_index:end_index]
* 슬라이싱, 2차원 ndarray : array[rindex_range, cindex_range]

* 팬시 인덱싱, 1차원 ndarray : array[[원소값들]]
* 팬시 인덱싱, 2차원 ndarray : array[[원소값들],[원소값들]]
* 불린 인덱싱 : arrary[조건식]

## ■ 넘파이 ndarray의 sort와 argsort
* sort()
 * np.sort(ndarray) : 인자로 들어온 원 행렬은 그대로 유지한 채 원 행렬의 정렬된 행렬을 반환
 * ndarray.sort() : 원 행렬 자체를 정렬한 형태로 반환하며 반환 값은 None
 * 기본적으로 오름차순으로 행렬 내 원소를 정렬하기 때문에 내림차순으로 정렬하기 위해서는 [::-1]을 적용하면 된다.
<br><br>
* 2차원 배열에서 axis 기반의 sort()

<br><br>
* argsort() : 원본 행렬 정렬 시 정렬된 행렬의 원래 인덱스를 필요로 할 때 np.argsort()를 이용한다. np.argsort()는 정렬 행렬의 원본 행렬 인덱스를 ndarray 형으로 반환한다.


___

* 행렬 정렬

```python
org_array = np.array([3, 1, 9, 5])
print('원본 배열 :', org_array, '\n')

# np.sort()로 정렬
sort_array1 = np.sort(org_array)
print('np.sort() 호출 후 반환된 정렬 배열 :', sort_array1)
print('np.sort() 호출 후 원본 배열', org_array, '\n')

# ndarray.sort()로 정렬
sort_array2 = org_array.sort()
print('org_array.sort() 호출 후 반환된 배열 :', sort_array2)
print('org_array.sort() 호출 후 원본 배열 :', org_array, '\n')
```

```python
sort_array1_desc = np.sort(org_array)[::-1]
print('내림차순으로 정렬 :', sort_array1_desc)
```

```python
array2d = np.array([[8, 12],
                    [7, 1]])

sort_array2d_axis0 = np.sort(array2d, axis = 0)
print('로우 방향으로 정렬 :\n', sort_array2d_axis0, '\n')

sort_array2d_axis1 = np.sort(array2d, axis = 1)
print('컬럼 방향으로 정렬 :\n', sort_array2d_axis1)
```

```python
org_array = np.array([3, 1, 9, 5])
sort_indices_desc = np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 원본 배열의 인덱스 :', sort_indices_desc)
```

```python
# np.argsort()를 어떻게 활용할 수 있을까?
import numpy as np

name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array = np.array([78, 95, 84, 98, 88])

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스 :', sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력 :', name_array[sort_indices_asc])
```

## ■ 넘파이 ndarray의 내적과 전치행렬
* np.dot(A, B) : 행렬 내적
* np.transpose(A) : 행렬 전치

___

* 행렬 내적

```python
A = np.array([[1, 2, 3], 
              [4, 5, 6]])
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])
dot_product = np.dot(A, B)
print('행렬 내적 결과 :\n', dot_product)
```

* 전치 행렬

```python
A = np.array([[1, 2],
              [3, 4]])
transpose_mat = np.transpose(A)
print('A의 전치 행렬 :\n', transpose_mat)
```
