### ■ 판다스 DataFrame의 변환, 컬럼 세트 생성 / 수정

* DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환<br>
 * 데이터프레임과 ndarray의 가장 큰 차이점 : 컬럼명의 여부


|변환 형태|설명|
|:---:|:---:|
|list -> DataFrame|df_list1 = pd.DataFrame(list, columns = col_name1)<br>와 같이 DataFrame 생성 인자로 리스트 객체와 매핑되는 컬럼명들을 입력|
|ndarray -> DataFrame|df_array2 = pd.DataFrame(array2, columns = col_name2<br>와 같이 DataFrame 생성 인자로 ndarray와 매핑되는 컬럼명들을 입력)|
|dict -> DataFrame|df_dict = pd.DataFrame(dict)와 같이 딕셔너리의 key로 컬럼명을, value를 리스트 형식으로 입력|
|DataFrame -> ndarray|DataFrame 객체의 values 속성을 이용하여 ndarray 변환|
|DataFrame -> list|DataFrame 객체의 values 속성을 이용하여 먼저 ndarray로 변환 후 tolist()를 이용하여 list로 변환|
|DataFrame -> dict|DataFrame 객체의 to_dict()를 이용하여 변환|

___

* 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

```python
import numpy as np

col_name1 = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)

print('array1 shape :', array1.shape, '\n')

df_list1 = pd.DataFrame(list1, columns = col_name1) # columns에는 무조건 list 형태로 넣어주어야 함
print('1차원 리스트로 만든 DataFrame :\n', df_list1, '\n')

df_array1 = pd.DataFrame(array1, columns = col_name1)
print('1차원 ndarray로 만든 DataFrame :\n', df_array1)
```

```python
# 3개의 컬럼명이 필요함
col_name2 = ['col1', 'col2', 'col3']

# 2행 * 3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1, 2, 3],
         [11, 12, 13]]
array2 = np.array(list2)

print('array2 shape :', array2.shape, '\n')

df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame :\n', df_list2, '\n')

df_array2 = pd.DataFrame(array2, columns = col_name2)
print('2차원 ndarray로 만든 DataFrame :\n', df_array2)
```

```python
# key는 칼럼명으로 매핑, value는 리스트형(또는 ndarray)
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame :\n', df_dict)
```

=> 리스트나 ndarray를 데이터프레임으로 변환할 때는 행 방향으로 데이터를 쌓고, 딕셔너리를 데이터프레임으로 변환할 때는 열 방향으로 데이터를 쌓는다.

### ■ 판다스 Index 개요
* 판다스 Index 객체는 RDBMS의 PK(Primary Key)와 유사하게 DataFrame, Series의 레코드를 고유하게 식별하는 객체이다. (하지만 판다스 Index는 별도의 컬럼값이 아니다.)
* DataFrame / Series 객체는 Index 객체를 포함하지만 DataFrame, Series 객체에 연산 함수를 적용할 때 index는 연산에서 제외된다. Index는 오직 식별용으로만 사용된다.
* index 객체를 추출하기 위해서는 index 속성을 사용한다.
 * Index 객체를 array로 받고 싶다면 values를 이용한다.
* 판다스 Index는 반드시 숫자형 값이 아니어도 되며, 고유한 값을 유지할 수 있다면 문자형/Datetime도 상관없다.
* DataFrame 및 Series에 reset_index() 메서드를 수행하면 새롭게 인덱스를 연속 숫자형으로 할당하며 기존 인덱스는 'index'라는 새로운 컬럼명으로 추가한다.
 * drop 옵션을 사용하면 기존의 인덱스를 유지할 지, 삭제할 지 설정할 수 있다.

___

### ■ DataFrame 인덱싱 및 필터링
* 1. [] : 컬럼 기반 필터링 또는 불린 인덱싱 필터링 제공
 * Numpy의 ndarray에서는 []에 위치 인덱스 값을 입력하여 해당 위치에 있는 값으로 ndarray를 반환
* 2. loc[], iloc[] : 명칭 / 위치 기반 인덱싱을 제공
* 3. 불린 인덱싱(boolean Indexing) : 조건식에 따른 필터링을 제공<br>
<br>
* 명칭 기반 인덱싱 : 컬럼의 명칭을 기반으로 위치를 지정하는 방식이며, loc[]를 사용한다.
* 위치 기반 인덱싱 : 기반 인덱싱은 0을 출발점으로 하는 가로축, 세로축 좌표 기반의 행과 열 위치를 기반으로 데이터를 지정하는 방식이며, 행, 열의 위치값으로 정수가 입력된다. (index를 이용하지 않는다.) iloc[]를 사용한다.
<br>
* 위치기반, 명칭기반 인덱싱 모두 사용할 필요없이 조건식을 [] 안에 기입하여 간편하게 필터링을 수행

=> 여러 집계함수를 사용할 때 agg내에 ''으로 감싸주지 않아도 된다.<br>
=> 여러 집계함수를 사용하는데 각 열 별로 다른 집계함수를 사용하면 ''으로 감싸주어야 한다.

```python
# agg 내의 인자로 들어가는 Dict 객체에 동일한 key 값을 가지는 두 개의 value가 있을 경우 마지막 value로 update됨
# 즉, 동일 컬럼에 서로 다른 aggregation을 가지면서 추가적인 컬럼 aggregation이 있을 경우 원하는 결과로 출력되지 않는다.

agg_format = {'Age':'max', 'Age':'mean', 'Fare':'mean'} # 딕셔너리에 동일한 두 개의 키 값이 있음 -> 마지막 것으로 업데이트
titanic_df.groupby('Pclass').agg(agg_format) # Age에 관한 연산이 두 가지이므로 마지막인 mean만 적용됨
```

```python
# 따라서 Named한 aggregation를 허용한다.

# 방법1
titanic_df.groupby(['Pclass']).agg(age_max = ('Age', np.max), age_mean = ('Age', 'mean'), fare_mean = ('Fare', 'mean'))
```

```python
# 방법2 : pd.NamedAgg를 이용
titanic_df.groupby('Pclass').agg(
    age_max = pd.NamedAgg(column = 'Age', aggfunc = 'max'),
    age_mean = pd.NamedAgg(column = 'Age', aggfunc = 'mean'),
    fare_mean = pd.NamedAgg(column = 'Fare', aggfunc = 'mean')
)
```

### ■ 판다스 람다식 적용하여 데이터 가공하기

* 파이썬 람다식 : 함수를 간단하게 한 줄로 적을 수 있음
* (ex) lambda x(매개변수): x ** 2(결과값)
* 판다스는 apply 함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능을 제공한다. 판다스의 경우 컬럼에 일괄적으로 데이터 가공을 하는 것이 속도 면에서 더 빠르나 복잡한 데이터 가공이 필요한 경우 어쩔 수 없이 apply lambda를 이용한다.

___

* 파이썬 lambda 식 기본

```python
# 함수식 이용
def get_square(a):
    return a ** 2

print('3의 제곱은', get_square(3))
```

```python
# lambda 식 이용
lambda_square = lambda x : x ** 2
print('3의 제곱은', lambda_square(3))
```

```python
a = [1, 2, 3]
squares = map(lambda x : x ** 2, a)
list(squares)
```

* 판다스 apply lambda 식 적용

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(10)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: 'Child' if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 위의 함수 정의가 헷갈린다면 따로 함수를 정의해주고 사용하면 됨
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else: cat = 'Elderly'
        
    return cat

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
